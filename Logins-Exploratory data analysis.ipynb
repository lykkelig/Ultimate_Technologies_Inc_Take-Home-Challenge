## Part 1 ‚Äê Exploratory data analysis

### The attached logins.json file contains (simulated) timestamps of user logins in a particular geographic location. Aggregate these login counts based on 15 minute time intervals, and visualize and describe the resulting time series of login counts in ways that best characterize the underlying patterns of the demand. Please report/illustrate important features of the demand, such as daily cycles. If there are data quality issues, please report them.

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data from logins.json file
df = pd.read_json('logins.json')
```

```python
# Examine the data
df.head()
```

```python
# Any data issues like null values?
df.isnull().values.any()
```

```python
# Convert login_time to a datetime 
df['login_time'] = pd.to_datetime(df['login_time'])
```

```python
# Make the login_time the index
df.set_index('login_time', inplace=True)
```

```python
# Arrange the login_time in 15 minute intervals and count logins
login_counts = df.resample('15T').size()
```

```python
# Plot the login_time
plt.figure(figsize=(12, 6))
login_counts.plot()
plt.title('Login Counts Over Time (15-minute Intervals)')
plt.xlabel('Time')
plt.ylabel('Login Counts')
plt.grid(True)
plt.show()
```

```python
# Arrange the login_time in hourly intervals and count logins
hourly_login_counts = df.resample('H').size()
```

```python
# Plot the hourly login_time
plt.figure(figsize=(12, 6))
hourly_login_counts.plot()
plt.title('Hourly Login Counts')
plt.xlabel('Time')
plt.ylabel('Login Counts')
plt.grid(True)
plt.show()
```

```python
# Arrange the login_time in daily intervals and count logins
daily_login_counts = df.resample('D').size()
```

```python
# Plot the daily login_time 
plt.figure(figsize=(12, 6))
daily_login_counts.plot()
plt.title('Daily Login Counts')
plt.xlabel('Time')
plt.ylabel('Login Counts')
plt.grid(True)
plt.show()
```
